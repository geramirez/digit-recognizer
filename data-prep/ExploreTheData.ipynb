```python
image = image_from_training_instance(instance_from_training_df(349))
image = image.astype('uint8')
edges = cv2.Canny(image,150,254,apertureSize = 3)

lines = cv2.HoughLines(edges,1,np.pi/180,200)
if lines:
    print("Lines detected:", lines)
    for rho,theta in lines[0]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))
        cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)
    show_image(img)
else:
    print("No lines detected")
```

## Translate image to top left

```python
contours, hierarchy = cv2.findContours(res,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

cnt = contours[0]

x,y,w,h = cv2.boundingRect(cnt)

show_image(res1, "Half Size")
rows,cols = res.shape
M = np.float32([[1,0,-x],[0,1,-y]])
dst = cv2.warpAffine(res1 ,M, (cols,rows))
show_image(dst, "Translated to top left")
```

```python
image = image_from_training_instance(instance_from_training_df(26761))
image = image.astype('uint8')


ret,thresh = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
show_image(thresh, "Image")

kernel = np.ones((2,2),np.uint8)
erosion = cv2.erode(thresh, kernel, iterations = 1)
show_image(erosion, "Eroded")

res = cv2.resize(erosion,None,fx=0.5, fy=0.5, interpolation = cv2.INTER_CUBIC)
res1 = res
res2 = res
show_image(res, "Half Size")

contours, hierarchy = cv2.findContours(res1,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
x,y,w,h = cv2.boundingRect(contours[0])

rows,cols = res.shape
M = np.float32([[1,0,-x],[0,1,-y]])
dst = cv2.warpAffine(res2 ,M, (cols,rows))
show_image(dst, "Translated to top left")
```
